```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The maximum x value for our chart will be 100
x_limit = 100

# List of values from 0 to 100
x_values = np.arange(x_limit)

# Create a random array of data of the same size as x_values
# that we will use for our y values
data = np.random.rand(len(x_values))
```

```python
# Create a scatter plot. The size of each point
# on our plot is determined by their x value
plt.scatter(x_values, data, marker="o", 
            facecolors="red", edgecolors="black",
            sizes=x_values, alpha=0.75)
```

```python
# The y limits of our scatter plot is 0 to 1
plt.ylim(0, 1)
```

```python
# The x limits of our scatter plot is 0 to 100
plt.xlim(0, x_limit)
```

```python
# Prints the scatter plot to the screen
plt.show()
```

```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
cwd = os.getcwd()
```

```python
inputFile1 = 'city_data.csv'
inputFile2 = 'ride_data.csv'
```

```python
csvpath1 = os.path.join(cwd, inputFile1)
csvpath2 = os.path.join(cwd, inputFile2)
```

```python
city_df = pd.read_csv(csvpath1)
ride_df = pd.read_csv(csvpath2)
city_df
```

```python
school_df = school_df.rename(columns = {'name': 'school name'})
len(school_df[school_df.duplicated(['school name'], keep=False)]) #checking for duplicates
```

```python
student_df = student_df.rename(columns = {'school': 'school name'})
len(student_df[student_df.duplicated(['Student ID'], keep=False)]) #checking for duplicates
```

```python
district_school_count = school_df.count()['school name']
```

```python
district_student_count = student_df.count()['Student ID']
```

```python
district_total_budget = school_df.sum()['budget']
```

```python
avg_math_score = student_df.mean()['math_score']
```

```python
avg_reading_score = student_df.mean()['reading_score']
```

```python
passed_math = student_df['math_score']>70
passed_math = np.count_nonzero(passed_math)
passed_math_pct = (passed_math / district_student_count) * 100
```

```python
passed_reading = student_df['reading_score']>70
passed_reading = np.count_nonzero(passed_reading)
passed_reading_pct = (passed_reading / district_student_count) * 100
```

```python
overall_passing_rate = (passed_math_pct + passed_reading_pct) / 2
```

```python
district_summary_df = pd.DataFrame([
                                    [district_school_count, district_student_count, district_total_budget, 
                                     avg_math_score, avg_reading_score, passed_math_pct, passed_reading_pct, 
                                     overall_passing_rate]
                                    ], columns = ['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score',
                                                    'Average Reading Score', '%Passing Math', '%Passing Reading', 
                                                      '%Overall Passing Rate'])

district_summary_df
```

```python
summary_df = pd.merge(school_df, student_df, how='outer', on='school name')
```

```python
school_summary1 = school_df[['school name', 'type', 'size', 'budget']]
school_summary1 = school_summary1.sort_values("school name")
school_summary1 = school_summary1.set_index("school name")
```

```python
school_summary1 = school_summary1.assign(bdgtPerStdnt = school_summary1['budget']/school_summary1['size'])
```

```python
school_summary3 = summary_df.groupby('school name')['reading_score', 'math_score'].mean()
```

```python
school_summary1 = school_summary1.assign(avg_reading_score = school_summary3['reading_score'])
school_summary1 = school_summary1.assign(avg_math_score = school_summary3['math_score'])
```

```python
passed_reading_score = summary_df['reading_score']>70
passed_math_score = summary_df['math_score']>70
```

```python
summary_df = summary_df.assign(passedReading = passed_reading_score)
summary_df = summary_df.assign(passedMath = passed_math_score)
```

```python
passed_reading_tests = summary_df.groupby('school name')['passedReading'].sum()
passed_school_reading_pct = (passed_reading_tests / school_summary1['size']) * 100
school_summary1 = school_summary1.assign(passed_reading_pct = passed_school_reading_pct)
```

```python
passed_math_tests = summary_df.groupby('school name')['passedMath'].sum()
passed_school_math_pct = (passed_math_tests / school_summary1['size']) * 100
school_summary1 = school_summary1.assign(passed_math_pct = passed_school_math_pct)
```

```python
overall_school_passing_rate = (passed_school_math_pct + passed_school_reading_pct) / 2
school_summary1 = school_summary1.assign(overall_passing_rate = overall_school_passing_rate)
school_summary1
```

```python
school_summary1 = school_summary1.sort_values("overall_passing_rate", ascending=0)
school_summary1.head(5)
```

```python
school_summary1 = school_summary1.sort_values("overall_passing_rate", ascending=1)
school_summary1.head(5)
```
